```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('train_loan.csv')

categorical_cols = ['Gender', 'Married', 'Dependents', 'Self_Employed']
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)


numerical_cols = ['LoanAmount', 'Loan_Amount_Term', 'Credit_History']
for col in numerical_cols:
    df[col].fillna(df[col].median(), inplace=True)


missing_values = df.isnull().sum()
missing_values[missing_values > 0]


```

```python
import numpy as np

def cap_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return np.clip(series, lower, upper)

df['ApplicantIncome'] = cap_outliers(df['ApplicantIncome'])
df['LoanAmount'] = cap_outliers(df['LoanAmount'])
```

```python
from sklearn.preprocessing import LabelEncoder

# Label Encoding
le = LabelEncoder()
df['Loan_Status'] = le.fit_transform(df['Loan_Status'])  # Y -> 1, N -> 0

# Convert 'Dependents' to numeric, replacing '3+' with 3
df['Dependents'] = df['Dependents'].replace('3+', 3).astype(int)

# One-Hot Encoding for remaining categorical variables
df_encoded = pd.get_dummies(df.drop(columns=['Loan_ID']), drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
num_cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']

df_encoded[num_cols] = scaler.fit_transform(df_encoded[num_cols])
```
